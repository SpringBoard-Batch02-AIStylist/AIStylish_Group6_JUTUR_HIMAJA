```python
import os
import numpy as np
import pandas as pd
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.layers import GlobalMaxPooling2D
from tensorflow.keras.models import Sequential
from numpy.linalg import norm
import matplotlib.pyplot as plt
from matplotlib.image import imread
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load the pre-trained ResNet50 model
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False
```

```python
# Create a sequential model with GlobalMaxPooling2D
model = Sequential([
    base_model,
    GlobalMaxPooling2D()
])
```

```python
def extract_features(image_path, model):
    img = image.load_img(image_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_expand_dim = np.expand_dims(img_array, axis=0)
    img_preprocess = preprocess_input(img_expand_dim)
    features = model.predict(img_preprocess).flatten()
    normalized_features = features / norm(features)
    return normalized_features
```

```python
# Collect image paths
image_folder_path = r"C:\Users\J UJWALA\Desktop\New folder\fashion product\fashion-dataset\fashion-dataset\images"
all_filenames = [os.path.join(image_folder_path, file) for file in os.listdir(image_folder_path)]
```

```python
# Create a DataFrame for easy manipulation
filenames_df = pd.DataFrame({'filename': all_filenames})
```

```python
# Randomly sample a subset of the dataset (e.g., 10% of the images)
sampled_df = filenames_df.sample(frac=0.1, random_state=42)  # Adjust the fraction as needed
subset_filenames = sampled_df['filename'].tolist()
```

```python
# Extract features from the sampled subset
image_features = np.array([extract_features(file, model) for file in subset_filenames])
```

```python
def get_cosine_similarity_recommendations(image_path, model, image_features, filenames, top_n=5):
    input_image_features = extract_features(image_path, model).reshape(1, -1)
    similarities = cosine_similarity(input_image_features, image_features)
    similarity_indices = similarities[0].argsort()[-top_n:][::-1]
    return similarity_indices
```

```python
# Specify the index of the sample image
sample_image_index = 369  # For example, use the 21st image in the subset
existing_image_path = subset_filenames[sample_image_index]

# Display the sample image
print(f"Displaying the sample image at index {sample_image_index}:")
sample_image = imread(existing_image_path)
plt.imshow(sample_image)
plt.axis('off')
plt.show()
```

```python
# Get similar image indices
similarity_indices = get_cosine_similarity_recommendations(existing_image_path, model, image_features, subset_filenames)

# Plotting recommendations
fig, axs = plt.subplots(1, 5, figsize=(20, 5))
for i in range(5):
    img = imread(subset_filenames[similarity_indices[i]])
    axs[i].imshow(img)
    axs[i].axis('off')
plt.show()
```
